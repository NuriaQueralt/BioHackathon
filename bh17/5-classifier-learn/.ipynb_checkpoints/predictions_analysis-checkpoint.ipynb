```python
import sys,os
import pandas as pd
import numpy as np
from sklearn import preprocessing
import matplotlib.pyplot as plt 
plt.rc("font", size=14)
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
import seaborn as sns
sns.set(style="white")
sns.set(style="whitegrid", color_codes=True)
```

```python
# path to data
path = os.getcwd() 

# read data
sys.path.insert(0, '.')
```

```python
predictions = pd.read_csv('predictions_mapped.csv', sep=',', header=None)
predictions = predictions.rename(
    columns={ 
        0: 'drug', 
        1: 'disease', 
        2: 'actual', 
        3: 'predicted', 
        4: 'error', 
        5: 'prediction'
    }
)

print(predictions.shape)
predictions.head(2)
```

```python
predictions = predictions[['drug', 'disease', 'predicted', 'prediction']]
predictions['predicted'] = predictions.predicted.apply(lambda x: 'true' if x.split(':')[1] == 't' else 'false')
predictions.head(2)
```

## Results exploration

```python
# check missing values
predictions.isnull().sum()
```

```python
sns.countplot(x= 'predicted', data= predictions, palette= 'hls')
plt.show()
```

```python
# barplot for true
true = predictions.query('predicted == "true"').copy()
sns.countplot(x= 'prediction', data= true)
plt.show()
```

```python
# barplot for true predictions
sns.countplot(x= 'prediction', data= true.query('prediction >= .98') )
plt.show()
```

```python
# barplot for false
false = predictions.query('predicted == "false"').copy()
sns.countplot(x= 'prediction', data= false)
plt.show()
```

```python
# barplot for false predictions
sns.countplot(x= 'prediction', data= false.query('prediction >= .98') )
plt.show()
```

```python
print('Percentage of true predictions (P >= .98): {}%'.format(round(
            100*len(true.query('prediction >= .98'))/len(true)))
     )
print('Percentage of false predictions (P >= .98): {}%'.format(round(
            100*len(false.query('prediction >= .98'))/len(false)))
     )
```

### Analysis of top 5 predictions
#### Include entity names alongside IDs

```python
# Mapping disease names
url = 'https://raw.githubusercontent.com/dhimmel/disease-ontology/75050ea2d4f60e745d3f3578ae03560a2cc0e444/data/slim-terms.tsv'
disease_df = pd.read_table(url)
disease_df = disease_df[['doid','name','pathophysiology']] 
disease_df['doid'] = ( disease_df
                  .doid
                  .apply(
                      lambda y: y.replace(':','_')
                        )
                          
             )
disease_df = disease_df.rename(columns={'doid': 'disease', 'name': 'disease_name', 'pathophysiology': 'disease_pathophysiology'})
disease_df.head(2)
```

```python
# Mapping drug names
url = 'https://raw.githubusercontent.com/dhimmel/drugbank/3e87872db5fca5ac427ce27464ab945c0ceb4ec6/data/drugbank-slim.tsv'
compound_df = pd.read_table(url)
compound_df = compound_df[['drugbank_id','name','categories']]
compound_df = compound_df.rename(columns={'drugbank_id': 'drug', 'name': 'drug_name', 'categories': 'drug_categories'})
compound_df.head(2)
```

```python
# mapping names to predictions dataframe
predictions = pd.merge(predictions,compound_df, how='left', on='drug')
predictions = pd.merge(predictions,disease_df,how='left',on='disease')
predictions.head(2)
```

#### Statistics of true predictions

```python
# sort true predictions
print('Percentage of true predictions (P = 1): {}%'.format(round(
            100*len(true.query('prediction == 1'))/len(true)))
     )
print('Number of true predictions (P = 1): {}'.format(
            len(true.query('prediction == 1')))
     )
```

```python
true = pd.merge(true,compound_df, how='left', on='drug')
true = pd.merge(true,disease_df,how='left',on='disease')
true.sort_values(by= 'prediction', ascending=False, inplace=True)
true.head(5)
```
